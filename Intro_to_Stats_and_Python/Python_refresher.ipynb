# Objects

```python
class Paperclip:
  ''' This is an object / class '''
  def __init__(self):
    # Define an instance variable
    self.material = 'aluminum'

  def get_material(self):
    return self.material

  def set_material(self, new_material):
    self.material = new_material
```

```python
# Access instance variables and functions using a .
new_paperclip = Paperclip()
print(new_paperclip.get_material())
```

```python
# Btw, functions are objects, too
# They are 'callable' objects
print(new_paperclip.get_material)
```

```python
# Variables can be accessed but not called
print(new_paperclip.material())
```

```python
# Inheritance
class GoldenPaperclip(Paperclip):
  def __init__(self):
    # Define an instance variable
    self.material = 'gold'

  def set_material(self, new_material):
    print('Material of this paperclip cannot be changed!')
```

```python
new_paperclip = GoldenPaperclip()
print(new_paperclip.get_material())
```

```python
class PaperclipMultiplier:
  def __init__(self):
    self.paperclip = Paperclip() # Use objects within other objects
    self.paperclip_collection = [self.paperclip]

  def multiply(self):
    self.paperclip_collection.append(Paperclip())

  def destroy_human_race(self):
    print("Don't distract me, I must multiply paperclips, nothing else!")
```
